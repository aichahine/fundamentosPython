```python
'''14/03/2026 Oitava aula '''
''' Modularização e tratamento de erros '''

''' 1. Crie uma função chamada apresentação que receba um nome, idade e curso
como parâmetro e imprima uma mensagem de apresentação para a sala, sendo que curso '''

''' O valor padrão é exibido quando o atributo não é invocado na função '''
def apresentacao(nome, idade, curso="FIC SENAI"):
  ''' Função que imprime uma mensagem de apresentação '''
  print(f"Olá, me chamo {nome}, tenho {idade} anos e estou no curso {curso}.")

''' Neste exemplo, um valor padrão foi definido para o atributo curso
e exibido na segunda invocação da função '''

nome_digitado = input("Digite seu nome: ")
idade_digitada = int(input("Digite sua idade: "))
curso_digitado = input("Digite o curso: ")

apresentacao(nome_digitado, idade_digitada, curso_digitado)
apresentacao(nome_digitado, idade_digitada)

```
