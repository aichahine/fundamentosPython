```python
'''07/03/2026 Sexta aula '''
''' Funções e Escopo '''

'''
Sintaxe da função
Argumentos (input) -> Função -> Valor do retorno (output)

def nomeDaFuncao(paramemetros)

Palavra-chave def
Nome da função
Parâmetro
Comando de retorno
Valor de retorno
'''

''' Definição da função '''
def fahrenheitToCelsius(temperatura):
  """
  Função que recebe como argumento uma temperatura
  em Fahrenheit e converte para Celsius
  """
  return ((temperatura-32)*(5/9))

''' Execução (invocação) da função '''
temperaturaCelsius = fahrenheitToCelsius(89)
print(f"{temperaturaCelsius:.2f}\n")

''' Invocando a função diretamente na impressão '''
print(fahrenheitToCelsius(92))

''' Retorna o valor do docstring '''
help(fahrenheitToCelsius)

def cumprimentar(nome):
  """
  Essa função irá imprimir um cumprimento com o nome passado por parâmetro
  """
  print(f"Olá, {nome}! Seja bem-vindo(a)!")

''' Invocando a função '''
cumprimentar("Aline")

''' Funções podem retornar um resultado ou imprimir um valor '''

''' Invocando uma função em um laço de repetição '''
''' Definição da função '''
def quadrado(numero):
  return numero*numero

tamanho = 6
numQuadrado = 0

''' Invocação da função '''
for i in range(tamanho):
  numQuadrado = quadrado(i)
  print(numQuadrado)

''' Exemplo '''

def soma(n1,n2):
  soma = n1+n2
  return soma

def subtracao(n1,n2):
  sub = n1-n2
  return sub

def multiplicacao(n1,n2):
  mult = n1*n2
  return mult

def divisao(n1,n2):
  if n2 == 0:
    print("Não é possível dividir um número por zero.")
  else:
    div = n1/n2
    return div

def potenciacao(n1,n2):
  pot = n1 ** n2
  return pot

def radiciacao(n1,n2):
  rad = n1 ** 0.5, n2 ** 0.5
  return rad


print("Bem-vindo à calculadora!\n")
num1 = float(input("Digite o primeiro número: "))
num2 = float(input("Digite o segundo número: "))
op = int(input("\nDigite:\n1 para somar\n2 para subtrair\n3 para multiplicar\n4 para dividir\n5 para potenciação\n6 para raiz quadrada\n"))
if op == 1:
  print(f"O valor da soma é: {num1} + {num2} = {soma(num1,num2)}")
elif op == 2:
  print(f"O valor da subtração é: {num1} - {num2} = {subtracao(num1,num2)}")
elif op == 3:
  print(f"O valor da multiplicação é: {num1} x {num2} = {multiplicacao(num1,num2)}")
elif op == 4:
  print(f"O valor da divisão é: {num1} / {num2} = {divisao(num1,num2)}")
elif op == 5:
  print(f"{num1} elevado a {num2} é = {potenciacao(num1,num2)}")
elif op == 6:
  print(f"A raiz quadrada de {num1} é = {radiciacao(num1,num2)}\n e a raiz quadrada de {num2} é = {radiciacao(num1,num2)}")
else:
  print("Operação inválida")
```
